```python
# Rename the column
df_u = df.rename(columns={'Annual Income (k$)': 'Income'}, inplace=False)

# Check if the column 'Income' exists now
print(df_u.columns)  # Verify column names

# Group by 'Genre' and calculate the mean of 'Income'
mean_income_by_genre = df_u.groupby('Genre')['Income'].mean()

# Print the result
print(mean_income_by_genre)
```
