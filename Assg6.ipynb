```python
'''
Download the Iris flower dataset or any other dataset into a
DataFrame. (e.g., https://archive.ics.uci.edu/ml/datasets/Iris).
Scan the dataset and give the inference as:
1. List down the features and their types (e.g., numeric,nominal) available in the dataset.
2.  Create a histogram for each feature in the dataset to
illustrate the feature distributions.
3.  Create a box plot for each feature in the dataset.
4.  Compare distributions and identify outliers.
'''

'''
1.Features in the Iris dataset:

sepal length (cm) → Numeric

sepal width (cm) → Numeric

petal length (cm) → Numeric

petal width (cm) → Numeric

species → Nominal (categorical: 'setosa', 'versicolor', 'virginica')
'''
import seaborn as sns

dataset=sns.load_dataset('iris')
dataset.head()
```

```python
dataset.describe()
```

```python
# Features and their types
print("Features and their types : ")
print(dataset.dtypes)
```

```python
#histogram for each feature
import matplotlib.pyplot as plt

fig,axes=plt.subplots(2,2,figsize=(16,9))
sns.histplot(x=dataset['sepal_length'],ax=axes[0,0])
sns.histplot(x=dataset['sepal_width'],ax=axes[0,1])
sns.histplot(x=dataset['petal_length'],ax=axes[1,0])
sns.histplot(x=dataset['petal_width'],ax=axes[1,1])
```

```python
#Box plot for each feature

fig,axes=plt.subplots(2,2,figsize=(16,9))
sns.boxplot(x='sepal_length',y='species',data=dataset,ax=axes[0,0])
sns.boxplot(x='sepal_width',y='species',data=dataset,ax=axes[0,1])
sns.boxplot(x='petal_length',y='species',data=dataset,ax=axes[1,0])
sns.boxplot(x='petal_width',y='species',data=dataset,ax=axes[1,1])
```
