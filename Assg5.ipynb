```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
titanic=sns.load_dataset('titanic')
titanic.head()
```

```python
'''
1. Use the inbuilt dataset 'titanic' as used in the
assignment#7. Plot a box plot for distribution of age with
respect to each gender along with the information about
whether they survived or not. (Column names: 'sex' and 'age')
2. Write observations on the inference from the above
statistics.
'''
# Create a boxplot for Age distribution
plt.figure(figsize=(8, 5))
sns.boxplot(x=titanic["age"], color="skyblue")

# Add title and labels
plt.title("Distribution of Age in Titanic Dataset", fontsize=14)
plt.xlabel("Age", fontsize=12)

# Show the plot
plt.show()

#maximum people are of below 40
```

```python
titanic.isnull().sum()
```

```python
titanic["age"].fillna(titanic["age"].median(), inplace=True)
titanic.isnull().sum()
```

```python
# Step 1: Compute the overall median age (ignoring NaN)
overall_median_age = titanic["age"].median()



#Replace age 0 with the overall median
titanic.loc[titanic["age"] == 0, "age"] = overall_median_age
```

```python
# Create a boxplot for Age distribution by Gender
plt.figure(figsize=(8, 5))
sns.boxplot(x="sex", y="age", data=titanic, palette={"male": "blue", "female": "pink"})

# Add title and labels
plt.title("Age Distribution by Gender in Titanic Dataset", fontsize=14)
plt.xlabel("Gender", fontsize=12)
plt.ylabel("Age", fontsize=12)

# Show the plot
plt.show()
#age of women is slightly lesser tham men
```

```python
# Create a boxplot for Age distribution by Gender and Survival Status
plt.figure(figsize=(10, 6))
sns.boxplot(x="sex", y="age", hue="survived", data=titanic, palette="Set2")

# Add title and labels
plt.title("Age Distribution by Gender and Survival Status in Titanic Dataset", fontsize=14)
plt.xlabel("Gender", fontsize=12)
plt.ylabel("Age", fontsize=12)

# Show the plot
plt.show()

#survival rate of women is greater than mens
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x="pclass", y="age", hue="survived", data=titanic, palette="Set2")

# Add title and labels
plt.title("Age Distribution by Class and Survival Status in Titanic Dataset", fontsize=14)
plt.xlabel("Passenger Class", fontsize=12)
plt.ylabel("Age", fontsize=12)

# Show the plot
plt.show()

#survival rate for first class is highest and degrade as per class
```
