```python
'''
1. Use the inbuilt dataset 'titanic'. The dataset contains 891
rows and contains information about the passengers who
boarded the unfortunate Titanic ship. Use the Seaborn library
to see if we can find any patterns in the data.
2. Write a code to check how the price of the ticket (column
name: 'fare') for each passenger is distributed by plotting a
histogram.
'''

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset=sns.load_dataset('titanic')  #Load titanic which is by default downloaded with seaborn library
dataset.head(10)
```

```python
#Distribution Plots
#1.Dist Plot
sns.displot(x=dataset['age'],bins=10,kde=True)
plt.xlabel('Age of Passengers')
plt.ylabel('Number of Passengers')
plt.title('Distribution of Age')

#2.Joint Plot
sns.jointplot(x=dataset['age'],y=dataset['fare'],kind='scatter')
sns.jointplot(x=dataset['age'],y=dataset['fare'],kind='hex')
```

```python

#3.Rug Plot
sns.rugplot(dataset['fare'])
plt.title('Rug Plot')
```

```python
#Catergorical Plots
#1.Bar Plot
sns.barplot(x='sex',y='age',data=dataset)
#sns.barplot(x='sex',y='age',data=dataset,estimator=np.std)
#it plots the standard deviation of age for each sex  basically showing how much variation there is in ages among males and females.
```

```python
#2.Count Plot
sns.countplot(x='sex',data=dataset,palette={'male': 'skyblue', 'female': 'lightpink'})
```

```python
#3.Box Plot
sns.boxplot(x='sex',y='age',data=dataset)
#sns.boxplot(x='sex',y='age',data=dataset,hue='survived')
```

```python
#4.Violin Plot
sns.violinplot(x='sex',y='age',data=dataset)
```

```python
#Advanced Plots
#1.strip plot
#sns.stripplot(x='sex',y='age',data=dataset,jitter=False)
sns.stripplot(x='sex',y='age',data=dataset,jitter=True)
```

```python
#2.swarm plot
sns.swarmplot(x='sex',y='age',data=dataset,hue='survived')
```

```python
#Matrix Plots
#1.Heat map
# Keep only numeric columns
numeric_data = dataset.select_dtypes(include='number')

# Now calculate correlation
cor = numeric_data.corr()

sns.heatmap(cor,annot=True)
```

```python
sns.clustermap(cor, annot=True,cmap='coolwarm', linewidths=0.5)
plt.title("Clustermap of Titanic Dataset")
plt.show()
```

```python
'''
1. Use the inbuilt dataset 'titanic'. The dataset contains 891
rows and contains information about the passengers who
boarded the unfortunate Titanic ship. Use the Seaborn library
to see if we can find any patterns in the data.
2. Write a code to check how the price of the ticket (column
name: 'fare') for each passenger is distributed by plotting a
histogram.
'''

# Example visualizations to find patterns:

# Survival rate by gender
sns.countplot(x='sex', hue='survived', data=dataset)
plt.title('Survival Count by Gender')
plt.show()

# Survival rate by class
sns.countplot(x='class', hue='survived', data=dataset)
plt.title('Survival Count by Passenger Class')
plt.show()

# Age distribution by survival
sns.histplot(data=dataset, x='age', hue='survived', multiple='stack')
plt.title('Age Distribution by Survival')
plt.show()
```

```python
# Plot histogram for ticket fare
sns.histplot(dataset['fare'], bins=30, kde=True)
plt.title('Fare Distribution')
plt.xlabel('Fare')
plt.ylabel('Number of Passengers')
plt.show()
#Summary in one line:
#"Most passengers traveled at low fares, with very few paying extremely high fares."
```
