```python
 #Data Formatting
df['new_writing_score']=df['writing_score'].astype("Int64")
df.dtypes
#df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler   #Min-Max Normalization

scaler = MinMaxScaler()
df[['new_writing_score']] = scaler.fit_transform(df[['new_writing_score']])
df.head(10)
```

```python
from sklearn.preprocessing import LabelEncoder   #Label Encoding

le = LabelEncoder()
df['gender_encoded'] = le.fit_transform(df['gender'])
df.head(10)
```

```python
from sklearn.preprocessing import OneHotEncoder  #one hot encoding

encoder = OneHotEncoder(sparse_output=False)   #sparse_output=False means give me a full matrix with 0s and 1s, not a compressed version.
enc_df = pd.DataFrame(encoder.fit_transform(df[['gender']]))
df=df.join(enc_df)
df.head(10)
```

```python
df_encoded = pd.get_dummies(df, columns=['gender'], drop_first=True)  #dummy variable encoding
print(df_encoded)
```
